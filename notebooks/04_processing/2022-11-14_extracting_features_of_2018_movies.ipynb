```python
from tmdbv3api import Movie
tmdb_movie = Movie() 
def get_genre(x):
    genres = []
    result = tmdb_movie.search(x)
    if not result:
      return np.NaN
    else:
      movie_id = result[0].id
      response = requests.get('https://api.themoviedb.org/3/movie/{}?api_key={}'.format(movie_id,tmdb.api_key))
      data_json = response.json()
      if data_json['genres']:
          genre_str = " " 
          for i in range(0,len(data_json['genres'])):
              genres.append(data_json['genres'][i]['name'])
          return genre_str.join(genres)
      else:
          return np.NaN
```

```python
# class to to scrape the data from the wikipedia page make api call to get the genre of the movie and then combine the data to get the final dataset

class extract_features:
    def __init__(self):
        pass
    def extract_from_wikipedia():
        link = "https://en.wikipedia.org/wiki/List_of_American_films_of_2019"
        df1 = pd.read_html(link, header=0)[2]
        df2 = pd.read_html(link, header=0)[3]
        df3 = pd.read_html(link, header=0)[4]
        df4 = pd.read_html(link, header=0)[5]
        df = pd.concat([df1, df2, df3, df4], ignore_index=True)
        return df
    

    def get_genre(df):
        genres = []
        for i in df['Title']:
            result = tmdb.Search().movie(query=i)
            if not result:
                genres.append(np.NaN)
            else:
                movie_id = result[0].id
                response = requests.get('https://api.themoviedb.org/3/movie/{}?api_key={}'.format(movie_id,tmdb.api_key))
                data_json = response.json()
                if data_json['genres']:
                    genre_str = " " 
                    for i in range(0,len(data_json['genres'])):
                        genres.append(data_json['genres'][i]['name'])
                else:
                    genres.append(np.NaN)
        return genres
    df['genres'] = df['Title'].map(lambda x: get_genre(str(x)))

    movies_for_2019 = df[['Title','Cast and crew','genres']]

    def get_directors(x):
        if ' (directors) ' in x:
            return x.split(' (directors) ')[0]
        else:
            return x.split(' (director/screenplay) ')[0]

    movies_for_2019['director_name'] = movies_for_2019['Cast and crew'].map(lambda x: get_directors(x))

    def get_actor1(x):
        return ((x.split('screenplay); ')[-1]).split(', ')[0])

    movies_for_2019['actor_1_name'] = movies_for_2019['Cast and crew'].map(lambda x: get_actor1(x))

    def get_actor2(x):
        if len((x.split('screenplay); ')[-1]).split(', ')) < 2:
            return np.NaN
        else:
            return ((x.split('screenplay); ')[-1]).split(', ')[1])

    movies_for_2019['actor_2_name'] = movies_for_2019['Cast and crew'].map(lambda x: get_actor2(x))

    def get_actor3(x):
        if len((x.split('screenplay); ')[-1]).split(', ')) < 3:
            return np.NaN
        else:
            return ((x.split('screenplay); ')[-1]).split(', ')[2])
    def final_dat():
        movies_for_2019['actor_3_name'] = movies_for_2019['Cast and crew'].map(lambda x: get_actor3(x))

        movies_for_2019 = movies_for_2019.rename(columns={'Title':'movie_title'})

        movies_for_2019 = movies_for_2019.loc[:,['movie_title', 'genres', 'director_name', 'actor_1_name', 'actor_2_name', 'actor_3_name']]

        movies_for_2019['actor_2_name'] = movies_for_2019['actor_2_name'].replace(np.nan, 'unknown')
        movies_for_2019['actor_3_name'] = movies_for_2019['actor_3_name'].replace(np.nan, 'unknown')

        movies_for_2019['movie_title'] = movies_for_2019['movie_title'].str.lower()

        movies_for_2019['comb'] = movies_for_2019['actor_1_name'] + ' ' + movies_for_2019['actor_2_name'] + ' '+ movies_for_2019['actor_3_name'] + ' '+ movies_for_2019['director_name'] +' ' + movies_for_2019['genres']

        return movies_for_2019
```
